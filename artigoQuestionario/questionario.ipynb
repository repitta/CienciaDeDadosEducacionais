```
import pandas as pd
import plotly.graph_objects as go
```

```
# table.csv has not missing values
table = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/Dados%20Demograficos.csv")
notas1A = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/notas_ECT2203_T01A_20192.csv", skiprows=11, sep= ";", encoding="ISO-8859-1",)
notas1B = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/notas_ECT2203_T01B_20192.csv", skiprows=11, sep= ";", encoding="ISO-8859-1")
notas1C = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/notas_ECT2203_T01C_20192.csv", skiprows=11, sep= ";", encoding="ISO-8859-1")
notas1D = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/notas_ECT2203_T01D_20192.csv", skiprows=11, sep= ";", encoding="ISO-8859-1")
notas3A = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/notas_ECT2203_T03A_20192.csv", skiprows=11, sep= ";", encoding="ISO-8859-1")
notas3B = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/notas_ECT2203_T03B_20192.csv", skiprows=11, sep= ";", encoding="ISO-8859-1")
notas3C = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/notas_ECT2203_T03C_20192.csv", skiprows=11, sep= ";", encoding="ISO-8859-1")
notas3D = pd.read_csv("https://github.com/repitta/CienciaDeDadosEducacionais/raw/master/artigoQuestionario/notas_ECT2203_T03D_20192.csv", skiprows=11, sep= ";", encoding="ISO-8859-1")
```

```
table = table.drop(columns=['Carimbo de data/hora', '2. Gênero:',
       '3. Idade: [Idade]', '4. Estado civil:', '5. Ocupação:',
       '6. Tem filhos ou alguém que depende de você? ',
       '7. Qual dos seguintes itens descreve melhor o nível mais alto de educação formal que você completou?',
       '8. Qual instituição você estudou antes de entrar na UFRN? ',
       '9. Quanto tempo, diariamente, você se dedica aos estudos ? ',
       '10. Quanto tempo, diariamente, você se dedica aos estudos de lógica de programação? ',
       '12. Qual o tempo médio de deslocamento utilizado diariamente até a UFRN?',
       '13. Em termos de conhecimento em progamação você se considera:',
       '14. Realizando atividades no computador você se considera:',
       '16. Qual é o nível mais alto de educação recebido por um dos seus pais?',
       '17. Você ver a disciplina de lógica de programação como diversão?',
       '19. Eu defino metas de curto prazo (diário ou semanal), bem como, metas de longo prazo.',
       '20. Eu mantenho um alto padrão de aprendizagem no meu curso.',
       '21. Eu mantenho um alto padrão de aprendizagem em lógica de programação.'])
table
```

```
 def process(df, colum):
    # divide media column into a range of values
    cut_points_notas = [-1,4,6]
    label_names_notas = ["pouco","muito"]
    df[colum] = pd.cut(df[colum],
                                 cut_points_notas,
                                 labels=label_names_notas)
    return df[colum]
 
```

```
copytable = table.copy()
copytable = copytable.drop(columns=["15. Qual o turno você realiza mais atividades de Lógica de Programação?","1. Matrícula"])
copytable["11. Você normalmente estuda no mesmo horário e local todos os dias."] = copytable["11. Você normalmente estuda no mesmo horário e local todos os dias."].astype(int)
co = copytable.columns
for i in co:
  copytable[i] = process(copytable, i)
copytable["matricula"] = table["1. Matrícula"] 

```

```
co = copytable.columns
for i in co:
  print( copytable[i].value_counts() )

```

```
notas = pd.concat([notas1A,notas1B,notas1C,notas1D,notas3A,notas3B,notas3C,notas3D])
notas = notas.loc[:,["Matrícula","Unnamed: 6"]]
notas = notas.dropna()
notas["Matrícula"] = notas["Matrícula"].astype(int)
notas = notas.rename(columns={"Matrícula": "matricula","Unnamed: 6":"notas"})
notas["notas"] = notas["notas"].str.replace(",",".",regex=False)
notas["notas"] = notas["notas"].astype(float)
notas
```

```
def processNotas(df, colum):
    # divide media column into a range of values
    cut_points_notas = [-1,6.9,11]
    label_names_notas = ["ruim","bom"]
    df[colum] = pd.cut(df[colum],
                                 cut_points_notas,
                                 labels=label_names_notas)
    return df[colum]
```

```
notas["notas"] = processNotas(notas,"notas")
```

```
merge = pd.merge(right = copytable, left = notas, on='matricula', how="inner")
merge
```

```
merge["notas_categoria"] = 1
merge.loc[merge["notas"] == "bom", "notas_categoria"] = 0 
```

```
titanic_df = merge.copy()

# Create dimensions
local_cat_dim = go.parcats.Dimension(values=titanic_df["18. Eu defino metas para a realização das minhas atividades."], label="Questão 18")

perguntas_dim = go.parcats.Dimension(values=titanic_df["22. Eu escolho sempre o mesmo local para estudar e evitar distrações."]	, label="Questão 22")

tempo_dim = go.parcats.Dimension(values=titanic_df['23. Eu procuro um lugar confortável para estudar.'], label="Questão 23")

comparativo_dim = go.parcats.Dimension(values=titanic_df['26. Eu preparo minhas perguntas antes de conversar com os monitores e professores.'], label="Questão 26")

tempoUniforme_dim = go.parcats.Dimension(values=titanic_df['29. Tento distribuir meu tempo estudando uniformemente todos os dias.'], label="Questão 29")

notas_dim = go.parcats.Dimension(values=titanic_df.notas_categoria, label="notas", categoryarray=[0, 1], ticktext=['bom', 'ruim'] )

# Create parcats trace
color = titanic_df.notas_categoria
colorscale = [[0, 'lightsteelblue'], [1, 'red']]

fig = go.Figure(data = [go.Parcats(dimensions=[local_cat_dim,perguntas_dim,tempo_dim, comparativo_dim,tempoUniforme_dim,notas_dim],
        line={'color': color, 'colorscale': colorscale},
        hoveron='color', hoverinfo='all',
        labelfont={'size': 18, 'family': 'Times'},
        tickfont={'size': 16, 'family': 'Times'},bundlecolors=True, 
        arrangement='freeform')])
fig.update_layout(width=800,height=500)

fig.show()
```

```
c = merge.columns
mergecopy=merge
for i in c:
  col = pd.Categorical(merge[i])
  mergecopy[i] = col.codes
mergecopy
```

```
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# feature selection
fsel_model = SelectKBest(score_func=chi2, k=10)
new_income = fsel_model.fit_transform(mergecopy,mergecopy.notas)

# only k=10 columns
new_income
```
